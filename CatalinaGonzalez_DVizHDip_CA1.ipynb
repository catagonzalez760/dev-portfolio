#

## Bars Chart

```python
def calc_median(arr):
    if len(arr) == 0:
        return None
    
    arr = arr.sort_values(ignore_index=True)

    n = len(arr)
    
    if n % 2 == 0:
       
        mid1 = arr[n // 2 - 1]
        mid2 = arr[n // 2]
        
        median = ((mid1 + mid2) / 2)
        
        return median
    
    else:
        median = arr[n // 2]
        return median
```

```python
older_games = vgsales[vgsales['Year_of_Release'] <= 2005]
newer_games = vgsales[vgsales['Year_of_Release'] > 2005]
```

```python
median_older = calc_median(older_games['EU_Sales'])
median_newer = calc_median(newer_games['EU_Sales'])
```

```python
print(f"Median of EU sales for games before 2005: {median_older}")
print(f"Median of EU sales for games after 2005: {median_newer}")
```

```python
if median_older > median_newer:
    print("Older games have a higher median EU sales.")
else:
    print("Newer games have a higher median EU sales.")
```

```python
labels = ['Older Games (<= 2005)', 'Newer Games (> 2005)']
medians = [median_older, median_newer]

plt.figure(figsize=(8, 6))
bars = plt.bar(labels, medians, color=['blue', 'orange'])
plt.title('Older games vs Newer games (EU_Sales)')
plt.xlabel('Year_of_Release')
plt.ylabel('Median EU_Sales (in millions)')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, 
             f'{height:.2f}', 
             ha='center', va='bottom')

plt.show()
```
