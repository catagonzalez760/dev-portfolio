# Students Academic Performance

##### This dataset is a database of various factors at the time of performance in students, this data contains 6,607 entries with the following 20 columns:

##### •Study Hours: Number of hours a student studies
##### •Attendance: Student’s attendance percentage.
##### •Parental Engagement: Level of involvement by parents in the student's education (e.g., "Low," "Medium").
##### •Resource Access: Access level to educational resources (e.g., "High," "Medium").
##### •Extracurricular: Participation in extracurricular activities ("Yes" or "No").
##### •Sleep Hours: Average sleep hours per day.
##### •Previous Scores: Past academic scores
##### •Motivation Level: Level of student motivation (e.g., "Low," "Medium").
##### •Internet Access: Access to the internet ("Yes" or "No").
##### •Tutoring Sessions: Number of tutoring sessions attended.
##### •Family_Income: Family income category (e.g., "Low," "Medium").
##### •Teacher Quality: Quality of teaching experienced by the student (e.g., "High," "Medium").
##### •School Type: Type of school attended (e.g., "Public," "Private").
##### •Peer Influence: Influence level from peers (e.g., "Positive," "Negative").
##### •Activity: Frequency of physical activity per week.
##### •Learning Disabilities: Presence of any learning disabilities ("Yes" or "No").
##### •Parental Education: Education level of parents (e.g., "High School," "College").
##### •Distance from Home: Distance of school from home (e.g., "Near," "Moderate").
##### •Gender: Gender of the student.
##### •Exam_Score: Student’s final exam score.

This dataset is suitable for analyzing how various demographic, social, and personal factors affect academic performance, providing insights for educational strategies.

```python
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
```

```python
import numpy as np
import mglearn 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

line = np.linspace(-3, 3, 1000).reshape(-1, 1)

for n_neighbors, ax in zip([1, 3,7], axes):
    
    reg = KNeighborsRegressor(n_neighbors = n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line), label="Model predictions")
    ax.plot(X_train, y_train, '^', c = mglearn.cm2(0), markersize = 8, label="Training data")
    ax.plot(X_test, y_test, 'v', c = mglearn.cm2(1), markersize = 8, label="Test data")
    ax.set_title(f"{n_neighbors} neighbor(s)\n train score:{reg.score(X_train, y_train):.2f} test score:{reg.score(X_test, y_test):.2f}")
    ax.set_xlabel("Study Hours")
    ax.set_ylabel("Exam_Score")
axes[0].legend(loc="best")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = Students_Academic_Performance[["Study Hours", "Attendance", "Parental Engagement", "Resource Access", "Extracurricular", "Sleep Hours", "Previous Scores", "Motivation Level", "Internet Access", "Tutoring Sessions", "Family_Income", "Teacher Quality",	"School Type", "Peer Influence", "Activity", "Learning Disabilities", "Parental Education", "Distance from Home", "Gender"]]  # independent colums
y = Students_Academic_Performance["Exam_Score"]  # targets variable

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,  random_state = 5) # 70% training and 30% test
```

```python
X.shape, y.shape, X_train.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_RF = RandomForestClassifier(n_estimators = 100)
clf_RF.fit(X_train, y_train)
y_pred = clf_RF.predict(X_test)
```

```python
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Rounded upto 2 decimal places
print( "Accuracy: {:.2f}".format(metrics.accuracy_score(y_test, y_pred)) )
```
